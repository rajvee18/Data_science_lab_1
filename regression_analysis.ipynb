```python
import numpy as np
from scipy import stats

X1 = np . array ([1 ,4 ,9 ,11 ,3 ,8 ,5 ,10 ,2 ,7 ,6] , dtype = float )
X2 = np . array ([8 ,2 , -8 , -10 ,6 , -6 ,0 ,12 ,4 , -2 , -4] , dtype = float )
y = np . array ([6 ,8 ,1 ,0 ,5 ,3 ,2 , -4 ,10 , -3 ,5] , dtype = float )
```

```python
n=len(y)

X=np.column_stack([np.ones(n),X1,X2])
k=X.shape[1]
```

```python
beta_hat = np.linalg.pinv(X) @ y
beta_hat
```

```python
y_hat = X @ beta_hat
y_hat
```

```python
e = y-y_hat
e
```

```python
ss_res = np.sum((y-y_hat)**2)
ss_res
```

```python
ss_reg = np.sum((y_hat-np.mean(y))**2)
ss_reg
```

```python
ss_tot = ss_reg+ss_res
ss_tot
```

```python
ms_res = ss_res/(n-k)
ms_reg = ss_reg/(k-1)
F=ms_reg/ms_res
F
```

```python
xtx_inv = np.linalg.pinv(X.T @ X)
se = np.sqrt(ms_res*np.diag(xtx_inv))
t_vals = beta_hat/se
t_crit = stats.t.ppf(0.025,8)
t_crit, t_vals
```
