```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Data

```python
x = np.array([2, 3, 4, 5, 6, 6, 7, 8, 9, 10, 3, 5, 7, 8, 9])
y = np.array([55, 60, 62, 68, 70, 72, 75, 78, 82, 88, 58, 66, 76, 80, 85])
```

## Calculating sample person coefficient

```python
r = np.sum((x-np.mean(x))*(y-np.mean(y)))/np.sqrt(np.sum((x-np.mean(x)**2))*np.sum((y-np.mean(y)**2)))
print(r)
```

## Boostrap resampling

```python
bootstrap_r = []
n=len(x)
for i in range(2000):
    indices = np.random.choice(range(n),size=n,replace=True)
    x_sample, y_sample = x[indices],y[indices]
    r = np.sum((x_sample-np.mean(x_sample))*(y_sample-np.mean(y_sample)))/np.sqrt(np.sum((x_sample-np.mean(x_sample)**2))*np.sum((y_sample-np.mean(y_sample)**2)))
    bootstrap_r.append(r)


```

## 95% confidence interval

```python
lower = np.quantile(bootstrap_r,0.025)
upper = np.quantile(bootstrap_r,0.975)
```

## Plotting

```python
plt.hist(bootstrap_r,bins=30)
plt.axvline(lower,color='green',label='95_percent_CI_lower')
plt.axvline(upper,color='red',label='95_percent_CI_upper')
plt.axvline(r,color='black',label = 'sample r')
plt.legend()
```
