```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

rng = np.random.default_rng(0)
```

```python
mu = np.array([3,70])
sigma1,sigma2,rho = 0.9,8,0.85
sigma = np.array([[sigma1**2,rho*sigma1*sigma2],[rho*sigma1*sigma2,sigma2**2]])
```

```python
samples = rng.multivariate_normal(mu,sigma,size=600)
x1,x2=samples[:,0],samples[:,1]

sns.kdeplot(x=x1,y=x2,fill=True,levels=10,cmap='coolwarm',thresh=0.05)
sns.scatterplot(x=x1,y=x2,s=10,color='black')
```
